## Procesamos los datos y consideramos fechas de interés
<p>Se considerarán como fecha de interés:</p>
<ul>

<p>Se considerarán años pandémicos: <b>2020, 2021, 2022</b></p>
<p>Se considerarán años de 'normalidad' o vuelta a esta: <b>2017, 2018, 2019, 2022</b></p>

```python
delitos = Delitos(path_to_data)
delitos.process_data()
delitos.date_of_interest = [[2019, 'Octubre', 'orangered'], [2020,'Marzo','orangered'] , [2022, 'Marzo','orangered']]
pandemic_years = [2020, 2021, 2022]
normal_years = [2017, 2018, 2019, 2022]
```

## Linea de tiempo

```python
fig, ax = plt.subplots(2,1,figsize=(15, 10)); fig.tight_layout(pad=5.0);
delitos.general_bar(folder_name, date_interest=True, ax=ax[0]);
delitos.general_plot(folder_name, date_interest=True, ax=ax[1]);
```
